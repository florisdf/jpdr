Measure the inference time of a separate detector and recognition network.

```python
from branched_models import get_timing_results
```

```python
import torch
```

```python
df_res = get_timing_results(lazy=True)
```

```python
import seaborn as sns
sns.set()

N_PROD = 'Number of products'
TIME = 'Time (s)'
DEVICE = 'Device'

df_res_cuda = df_res[df_res[DEVICE] == 'cuda']
df_res_cpu = df_res[df_res[DEVICE] == 'cpu']

sns.relplot(data=df_res_cuda,
            x=N_PROD, y=TIME,
            hue='Branch layer', kind='line', style='Branch layer', col=DEVICE)
```

```python
sns.relplot(data=df_res_cpu,
            x=N_PROD, y=TIME,
            hue='Branch layer', kind='line', style='Branch layer', col=DEVICE)
```
